````python
# scikit-learn 라이브러리에서 iris(붓꽃) 데이터 셋을 임포트
from sklearn.datasets import load_iris
iris = load_iris()

# X(독립변수를 iris.data에 저장되어있는 것을 변수로 따로 저장)
# y(종속변수를 iris.target에 저장되어있는 것을 변수로 따로 저장)
X = iris.data
y = iris.target
````

````python
# 신경망은 로지스틱 회귀를 내부적으로 사용함
# 로지스틱 회귀는 좌표공간에 그려지기 때문에 표준화를 사용했을 경우 장점이 많음
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
````

````python
# 위에 선언했던 독립변수, 종속변수를 train_test_split에 넣음.
# 각 데이터셋에 랜덤성을 유지하기 위해 층화추출인 stratify 파라미터를 사용함.
# random_state는 아무값을 넣어도 상관없음. 다만 재현성을 위해서 임의의 값을 넣어서 동작함.
# 여기에서는 관례적으로 사용되는 42를 사용
X_train, X_test, y_train, y_test = train_test_split(
    X, y, stratify=y, random_state=42
)
````

````python
# 데이터셋의 평균과 표준편차를 이용하여 표준화시키는 StandardScaler를 사용함
# fit_transform을 활용하여 train set의 평균과 표준편차로 test set의 정규화를 해줌
# train과 test를 나누어 모델이 실제 상황과 유사한 환경을 맞추어줌
scaler = StandardScaler()
X_train_zs = scaler.fit_transform(X_train)
X_test_zs = scaler.transform(X_test)
````

````python
# scikit-learn 라이브러리에서 neural_network를 위한 다중퍼셉트론 분류모델을 가지고 옴
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
````

````python
# 다중퍼셉트론 클래스를 iris_nn 변수에 저장
iris_nn = MLPClassifier()
# iris_nn.fit으로 X_train데이터를 표준화한 것을 모델에 학습시킴
iris_nn.fit(X_train_zs, y_train)
# iris_nn.predict에 X_test데이터를 표준화한 것을 넣어 예측값을 받아옴
pred = iris_nn.predict(X_test_zs)

# 
print(accuracy_score(y_test, pred))
print(confusion_matrix(y_test, pred))
````

````python
# 다중퍼셉트론 클래스를 iris_nn 변수에 저장
iris_nn = MLPClassifier((10, 50, 10))
# iris_nn.fit으로 X_train데이터를 표준화한 것을 모델에 학습시킴
iris_nn.fit(X_train_zs, y_train)
# iris_nn.predict에 X_test데이터를 표준화한 것을 넣어 예측값을 받아옴
pred = iris_nn.predict(X_test_zs)

# 
print(accuracy_score(y_test, pred))
print(confusion_matrix(y_test, pred))
````

````python
# 다중퍼셉트론 클래스를 iris_nn 변수에 저장
iris_nn = MLPClassifier((10, 50, 10), random_state=42)
# iris_nn.fit으로 X_train데이터를 표준화한 것을 모델에 학습시킴
iris_nn.fit(X_train_zs, y_train)
# iris_nn.predict에 X_test데이터를 표준화한 것을 넣어 예측값을 받아옴
pred = iris_nn.predict(X_test_zs)

# 
print(accuracy_score(y_test, pred))
print(confusion_matrix(y_test, pred))
````

## 정리

- MLPClassifier는 히든레이어로 기본값을 (100, )으로 가짐.
- 사용자 지정 값으로 주려면 튜플을 주면 됨. 
- random_state를 지정하지 않으면 돌릴 때마다 다른 결과값이 반환됨
- max_iter를 주지 않았을 경우 수렴하지 않기 때문에 경고가 발생함

```py
iris_nn = MLPClassifier((10, 50, 10), random_state=42)
```

````python
iris_nn = MLPClassifier((10, 50, 10), random_state=42, max_iter=1000)
iris_nn.fit(X_train_zs, y_train)
pred = iris_nn.predict(X_test_zs)

print(accuracy_score(y_test, pred))
print(confusion_matrix(y_test, pred))
````

- accuracy_score는 변화가 없더라도 max_iter를 반복할수록 오차가 줄어듦
- 더 적은 오차로 예측을 할 수 있다는 뜻
- 더 적은 오차로 예측이 항상 더 좋은 예측은 아님, loss가 너무 줄면 train 데이터에 과적합될 가능성이 높아짐

````python
# 확률값으로 예측할 수 있음
# 99.95%
# 0.0195%
# 0.0305%
# 한 행의 합이 100%가 됨 
# setosa일 확률이 99.94%라는 뜻

# 두 번째 행을 보면
# 0.58%
# 91.4%
# 7.944%
# versicolor가 될 확률이 91.4%
iris_nn.predict_proba(X_test_zs)
````
